```python
Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
ans:
    
  Here are eight Seaborn plots and their uses in plain English:

**1. Scatter Plot:** Like throwing darts at a board, shows how two things relate to each other. Dots closer together mean stronger connection.
Useful for checking if one variable goes up when another does.

**2. Bar Plot:** Think of stacked candy bars, each bar shows how much of something belongs to a category. Great for comparing quantities 
across different groups.

**3. Histogram:** Imagine buckets collecting your data, shows how many fall in each range. Helpful for understanding data spread
and if it's bunched up or spread out.

**4. Box Plot:** Like a fancy mustache, shows the middle half of your data, the spread, and outliers. Useful for quickly comparing groups and 
spotting unusual values.

**5. Violin Plot:** Think of a squashed violin showing data density. Similar to box plots but also shows distribution shape. Good for comparing 
both center and spread across groups.

**6. Heatmap:** Like a colorful temperature map, shows how two things vary together. Darker areas mean more of something happening. Great for 
exploring relationships between many variables at once.

**7. Pair Plot:** Think of a grid of scatter plots, shows how each pair of variables relates. Useful for seeing all pairwise relationships in your 
data in one place.

**8. Facet Grid:** Imagine splitting your data into smaller plots, then making the same plot for each group. Helps compare trends within different 
categories of your data.

These are just eight examples, Seaborn has many more plots! Remember, the best plot depends on your data and what you want to learn from it.


    
    
    
    
    
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
ans:
    import seaborn as sns
import matplotlib.pyplot as plt

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create the line plot using relplot for multiple lines
sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    kind="line",
)
plt.show()

    
Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.
ans:
    
  
import seaborn as sns
import matplotlib.pyplot as plt

# Load the titanic dataset
titanic = sns.load_dataset("titanic")

# Plot two box plots
plt.figure(figsize=(12, 6))

# Box plot for 'age' based on 'pclass'
plt.subplot(1, 2, 1)
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Box plot of Age based on Pclass')

# Box plot for 'fare' based on 'pclass'
plt.subplot(1, 2, 2)
sns.boxplot(x='pclass', y='fare', data=titanic)
plt.title('Box plot of Fare based on Pclass')

# Show the plots
plt.show()




Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
ans:
    
import seaborn as sns
import matplotlib.pyplot as plt
diamonds = sns.load_dataset("diamonds") 
sns.displot(x = diamonds.carat , y=diamonds.price , data=diamonds ,hue='cut')
    
    
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
ans:
    
    import seaborn as sns
    import matplotlib.pyplot as plt
    iris = sns.load_dataset("iris")
    sns.pairplot(iris)
    
    
    
    
    
    
Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
ans:
    import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights_data = sns.load_dataset("flights")

# Pivot the data to create a matrix suitable for a heatmap
flights_pivot = flights_data.pivot_table(index='month', columns='year', values='passengers')

# Create a heatmap
sns.heatmap(flights_pivot, cmap='Blues', annot=True, fmt='d', linewidths=.5)

# Set the title and labels
plt.title('Passenger Count by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

# Display the heatmap
plt.show()





  ####  
import seaborn as sns
    import matplotlib.pyplot as plt
    flights = sns.load_dataset("flights")
    sns.heatmap(flights)
    ######
    
    
    
    
    
    
    
    
    
    
    
    
```
